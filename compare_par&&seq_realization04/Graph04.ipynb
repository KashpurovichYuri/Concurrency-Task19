```python
def AxisSettings(ax):
    # majorLocator = MultipleLocator(5)
    xaxminorLocator = AutoMinorLocator(n=10)
    yaxminorLocator = AutoMinorLocator(n=5)
    
    # OX
    xax = ax.get_xaxis()
    xlabels = xax.get_ticklabels()
    xlines = xax.get_ticklines()
    for label in xlabels:
        label.set_color('black')  # axis label color
        label.set_rotation(0)  # rotation of labels of divisions of the axis
        label.set_fontsize(18)  # font size of labels of divisions of the axis
    for line in xlines:
        line.set_color('black')  # set the color of the dividing line
        line.set_markersize(3.5)  # set the length of the dividing line
        line.set_markeredgewidth(0.75)  # set the thickness of the division line
    # xax.set_major_locator(majorLocator)
    xax.set_minor_locator(xaxminorLocator)
    xax.set_minor_formatter(NullFormatter())
    
    # OY
    yax = ax.get_yaxis()
    ylabels = yax.get_ticklabels()
    ylines = yax.get_ticklines()
    for label in ylabels:
        label.set_color('black')  # axis label color
        label.set_rotation(0)  # rotation of labels of divisions of the axis
        label.set_fontsize(18)  # font size of labels of divisions of the axis
    for line in ylines:
        line.set_color('black')  # set the color of the dividing line
        line.set_markersize(3.5)  # set the length of the dividing line
        line.set_markeredgewidth(0.75)  # set the thickness of the division line
    # yax.set_major_locator(majorLocator)
    yax.set_minor_locator(yaxminorLocator)
    yax.set_minor_formatter(NullFormatter())
```
